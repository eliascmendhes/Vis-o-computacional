Pixel é a menor unidade de informação que compõe uma imagem digital. Em uma escala de imagem RGB o pixel é representado por três números de 8 bits associados com as cores R, G, B que correspodem as cores Red, Green e Blue respectivamente. Como são números de 8bits a faixa de valores é de 0 a 255.
